```python
import numpy as np
import pandas as pd

# Create student dataset
data = {
    "Roll_No": np.arange(1, 11),
    "Marks": [85, 90, np.nan, 78, 88, np.nan, 92, 70, 80, 75],
    "Attendance": [90, 85, 88, np.nan, 92, 80, np.nan, 75, 89, 84],
    "Gender": ["M", "F", "F", "M", np.nan, "F", "M", "M", "F", np.nan]
}

f = pd.DataFrame(data)

x=f["Gender"].isnull().sum()
print(x)

f["Gender"].fillna(f["Gender"].mode()[0],inplace=True)

avg=(f["Marks"].sum())/10
avg

(f["Gender"]=="F").sum()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.bar(f["Gender"].value_counts().index,f["Gender"].value_counts().values,color=["b","y"])
plt.title('graph')
plt.xlabel('gender of student')
plt.ylabel('number of student')
plt.show()
```
